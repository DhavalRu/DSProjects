<h1><center>Craigslist Web Scraper</center></h1>

This notebook grabs the latest rent and condo property data from Craigslist. It will loop through all the pages and store the following information: Date/time, ID, Link to the detailed posting, Price, Square Feet, Address/Intersection, and Number of Bedrooms

```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup

#fRent = open(fileRent, 'w')
#fBuy = open(fileBuy, 'w')

def scraping_process(rent_or_buy):
    buy_url = 'https://toronto.craigslist.ca/search/rea?bundleDuplicates=1&availabilityMode=0&housing_type=1&housing_type=2&housing_type=5&sale_date=all+dates'
    rent_url = 'https://toronto.craigslist.ca/search/apa?bundleDuplicates=1&availabilityMode=0&housing_type=1&housing_type=2&housing_type=5&sale_date=all+dates'

    fileRent = 'Data/Craigslist_rent.csv'
    fileBuy = 'Data/Craigslist_buy.csv'
    if rent_or_buy == 'rent':
        url = rent_url
        file = fileRent
    else:
        url = buy_url
        file = fileBuy
    f = open(file, 'w')
    headers = "Datetime,ID,Link,Price,SqrFeet,Hood,Housing\n"
    f.write(headers)
    while url:
        print ("Web Page: ", url)
        client = uReq(url)
        html = client.read()
        client.close()
        a_soup = soup(html, 'html.parser')
        containers = a_soup.findAll("p", {"class": "result-info"})
        for container in containers:
            datetime = container.time['datetime']
            ID = container.a['data-id']
            link = container.a['href']
            # More detailed scraping, commented out since details not too interesting, also web scraping took over an hour
            '''
            # Same process as before to get some details for each listing
            details_Client = uReq(link)
            details_html = details_Client.read()
            details_Client.close()
            details_soup = soup(details_html, 'html.parser')
            details_containers = details_soup.find_all("div", class_="mapAndAttrs")
            try:
                lat = details_containers[0].find(class_ = 'viewposting')['data-latitude'] # Latitude
                long = details_containers[0].find(class_ = 'viewposting')['data-longitude'] # Longitude
            except:
                lat = 0
                long = 0
            try:
                if 'google map' not in details_containers[0].div.find(class_ = 'mapaddress').get_text():
                    addr = details_containers[0].div.find(class_ = 'mapaddress').get_text().replace(",", "") # Address
                else:
                    addr = 'N/A'
            except:
                addr = 'N/A'
            try: 
                bed_bath = details_containers[0].find(class_= 'shared-line-bubble').get_text() # Bed and bath
                bed = details_containers[0].find(class_= 'shared-line-bubble').get_text().split('/')[0]
            except:
                bed_bath = 'N/A'
                bed = 'N/A' ''' 
            try:
                price = container.find(class_ = 'result-price').get_text()
            except:
                price = 0
            try: # Try to see if sqr feet is available
                sqr_feet = 0
                for i in container.find(class_ = 'housing').get_text().strip().split('-'):
                    if 'ft' in i:
                        sqr_feet = i.strip()
            except:
                sqr_feet = 0
            try:
                hood = container.find(class_ = 'result-hood').get_text().replace(",", "")
            except:
                hood = 0
            try:
                bed = 0
                for i in container.find(class_ = 'housing').get_text().strip().split('-'):
                    if 'br' in i:
                        bed = i.strip()
            except:
                bed = 0
           # f.write(str(datetime) + "," + str(ID) + "," + str(link) + "," + str(price).replace("$", "") + "," + str(sqr_feet) + "," + str(hood) 
           #             + "," + str(lat) + "," + str(long) + "," + str(addr) + "," + str(bed_bath) + "," + str(bed) + '\n')
            f.write(str(datetime) + "," + str(ID) + "," + str(link) + "," + str(price).replace("$", "") + "," + 
                    str(sqr_feet) + "," + str(hood) + "," + str(bed) + '\n')
        nextlink = a_soup.find("link", rel="next")
        url = False
        if (nextlink):
            url = nextlink['href']
    f.close()
    return

scraping_process('rent')
scraping_process('buy')
```
